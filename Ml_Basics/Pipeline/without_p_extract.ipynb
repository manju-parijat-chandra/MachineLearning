```python
import pickle
import numpy as np
import pandas as pd
```

```python
# Decoding to Decision tree Object
# Load Pickel

ohe_sex = pickle.load(open('../../PickelFiles/wp_ohe_sex.pkl','rb'))
ohe_embarked = pickle.load(open('../../PickelFiles/wp_ohe_embarked.pkl','rb'))
clf = pickle.load(open('../../PickelFiles/wp_clf.pkl','rb'))
```

```python
# Prediction on user data

# Assume user input
# Pclass/gender/age/SibSp/Parch/Fare/Embarked
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1,7) # Sklearn Requires Matrix -> reshape
```

```python
# We have to apply all the preprocessing steps Like we did before
# Order Should Be same 

# Reshape is required ->  The encoder needs a 2D "matrix" shape. sex-> ['male'] (Before Encoding) to [[0,1]](after Encoding)

# test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
temp_sex_df = pd.DataFrame(test_input[:, 1].reshape(1, 1), columns=['Sex'])
test_input_sex = ohe_sex.transform(temp_sex_df) # To match the Column name 

test_input_embarked = ohe_embarked.transform(pd.DataFrame(test_input[:,-1].reshape(1,1)))
test_input_age = test_input[:,2].reshape(1,1)

# -> Encoded the Input column -> Ready for Prediction

# Concatenate as before
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```

```python
# Prediction

# Convert your concatenated array to float (Numpy array)
# This is usually what the model expects if it was trained on 'transformed' data
test_input_final = test_input_transformed.astype(float)

# Predict directly
prediction = clf.predict(test_input_final)

# 3. Output result
print("Result:", "Survived" if prediction[0] == 1 else "Not Survived")
```
