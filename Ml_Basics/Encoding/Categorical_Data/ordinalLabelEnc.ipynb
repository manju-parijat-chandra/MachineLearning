```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../../DataBox/customer.csv')
```

```python
# Categorical Columns are -> gender, review, education -> (In Input columns), purchased -> (in output column)
# Nominal categorical (No Order Is possible) -> Gender (M/F), Purchased (Yes/No)
# Ordinal Categorical (Order is Possible)    -> Review (Good/Average/Poor), Education (School, UG, PG)

# Review and Education  ->  Ordinal Encoding
# purchased             -> Label Encoding

# As Ordinal Enc. is For Input Categorical Column Only

df = df.iloc[:,2:]

# First Train Test Split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1], test_size=0.2)

# Create Object
# Feed Order Of the Values
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']]) # If Order is not given -> Randomly Assigned

oe.fit(X_train)

X_train = oe.transform(X_train)
X_test = oe.transform(X_test)

# Label Enc. As it is For Output Column Only
# encoded Target Labels 0 to n_classes - 1

# Create Object of Label Encoder
# we do not decide order -> Randomly Assigned
le = LabelEncoder()

# Fit on Train set
le.fit(y_train)

# Transform Both Train And Test Sets
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
