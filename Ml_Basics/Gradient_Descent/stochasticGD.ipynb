```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split 
```

```python
# IO
X,y = load_diabetes(return_X_y=True)

# TTS
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Load Model
reg = LinearRegression()
reg.fit(X_train,y_train)

# Check Coefficients
# B1, B2 ... Bm
print(reg.coef_)
# B0
print(reg.intercept_)

# Do Prediction For Varification
y_pred = reg.predict(X_test)
# R2 Score
print(r2_score(y_test,y_pred))
```

```python
class StochasticGD:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            # Working on one row at once
            
            # Update total_row times per Epoch 
            for j in range(X_train.shape[0]):
                
                # Pick one Random Row
                idx = np.random.randint(0,X_train.shape[0])
                
                # only one y_hat Rather Than Mean of All as in BatchGD
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                
                intercept_der = -2 * (y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                
                # Update all Beta with only one row -> 1 * Num_Col
                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
        # Total Check -> Epoch * Total_Row * (1*Num_Col)
        # Same as GD but this approach does small change on every step. -> Smooth convergence
        # And It check with Highest Probability Value Rows Rather than checking each and every row
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
gdr = StochasticGD(epochs=50,learning_rate=0.01)

gdr.fit(X_train,y_train)

y_pred = gdr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
