```python
from sklearn.datasets import make_regression
import numpy as np

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)

plt.scatter(X,y)
```

```python
# Apply OLS (Ordinary Least Square) (Formula Method)

reg = LinearRegression()
reg.fit(X,y)

m = reg.coef_
print('m ->' , m)

actual_b = reg.intercept_

print('We Have to Reach b -> ', actual_b , ' by Starting Randomly')
```

```python
# Regression Line -> We have to reach Here

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)

# Plot Graph How Close We have Started
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 1
# Objective -> Calculate b new
# we need slope 

# Let value of m, lr, b old
m = 78.35
lr = 0.1
b = 0

# Slope at b = 0
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print(loss_slope)

step_size = loss_slope*lr
print(step_size)

# Calculating the new intercept
b = b - step_size
print(b)

b1 = b
```

```python
# Check how Close we have moved

y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b1 = {}'.format(b1))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
# Objective -> Calculate b new
# we need slope 

# Let value of m, lr, b old
m = 78.35
lr = 0.1

# Slope at b = 0
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print(loss_slope)

step_size = loss_slope*lr
print(step_size)

# Calculating the new intercept
b = b - step_size
print(b)

b2 = b
```

```python
# Check how Close we have moved

y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b2 = {}'.format(b2))
plt.plot(X,y_pred1,color='#A3E4D7',label='b1 = {}'.format(b1))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
# Objective -> Calculate b new
# we need slope 

# Let value of m, lr, b old
m = 78.35
lr = 0.1

# Slope at b = 0
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print(loss_slope)

step_size = loss_slope*lr
print(step_size)

# Calculating the new intercept
b = b - step_size
print(b)

b3 = b
```

```python
# Check how Close we have moved

y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b3 = {}'.format(b3))
plt.plot(X,y_pred2,color='#A3E4D7',label='b2 = {}'.format(b2))
plt.plot(X,y_pred1,color='#A3E4D7',label='b1 = {}'.format(b1))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()

# Observation -> b value is getting closer to Actual b
```

Do These steps in Loop

```python
# Loss Slope is high means -> small change is b makes huge impact on Loss function
b = -100
m = 78.35
lr = 0.05

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)
  print(f'Iteration {i} Loss is {loss_slope} -> b_New ->' , b)
  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)

# Even we start with worst value b -> 100
# we Can See how we converged to Actual b value
```
