```python
# Problem 1 

# Outlook-Sunny, Temp-Hot, Humidity-High, wind-Weak
# Play -> Yes/No

# Solution (Prediction)

# P(Yes | Sunny, Hot, High, Weak) = P(Sunny | Yes) * P(Hot | Yes) * P(High | Yes) * P(Weak | Yes) * P(Yes) / 2 (Divide for exact Probability)
# P(No | Sunny, Hot, High, Weak) = P(Sunny | No) * P(Hot | No) * P(High | No) * P(Weak | No) * P(No) / 2 (Can Skip divide as it is always proportional)

# These Two are compared Using maximum Posteriori rule
```

```python
# Problem 2

# Outlook - OverCast, Temp - Cold, Humidity - Low, wind - Weak
# Play -> Yes/No

# Solution (Prediction)

# P(Yes | Overcast, Cool, Low, Weak) = P(Overcast | Yes) * P(Cool | Yes) * P(Low | Yes) * P(Weak | Yes) * P(Yes) / p(xi) (Divide for exact Probability)
# P(No | Overcast, Cool, Low, Weak) = P(Overcast | No) * P(Cool | No) * P(Low | No) * P(Weak | No) * P(No) * P(No) / p(xi) (Can Skip divide as it is always proportional)

# These Two are compared Using maximum Posteriori rule
```

```python
# Training Stage

# Create a Lookup Table (Dictonary)

# Calculate all the probability (Of all Combination) and stores as a Dictonary
```

```python
# Calculating all the probabilities

# P(Y)

py_dict = df['Play'].value_counts(normalize=True).to_dict()

# P(X | Y)
pd.crosstab(df['Outlook'], df['Play'],normalize='columns').stack().to_dict()

px_dict = {}
for i in ['Outlook','Temp',	'Humidity',	'Wind']:
  px_dict.update(pd.crosstab(df[i], df['Play'],normalize='columns').stack().to_dict())
```

```python
# testing Stage

# problem 1.  
# P(Yes | Sunny, Hot, High, Weak) = P(Sunny | Yes) * P(Hot | Yes) * P(High | Yes) * P(Weak | Yes) * P(Yes)
# P(No | Sunny, Hot, High, Weak) = P(Sunny | No) * P(Hot | No) * P(High | No) * P(Weak | No) * P(No)


# Query Point
outlook, temp, humidity, wind = 'Sunny', 'Hot', 'High', 'Weak'


play = 'Yes'
p_yes_sunny_hot_high_weak = px_dict[(outlook, play)]*px_dict[(temp, play)]*px_dict[(humidity, play)]*px_dict[(wind, play)] * py_dict['Yes']

play = 'No'
p_no_sunny_hot_high_weak = px_dict[(outlook, play)]*px_dict[(temp, play)]*px_dict[(humidity, play)]*px_dict[(wind, play)] * py_dict['No']
```

```python
print('Yes - ', p_yes_sunny_hot_high_weak)
print('No - ', p_no_sunny_hot_high_weak)

# Observation ->
# Probability of No is More than Yes, So There Won't be any play
```
