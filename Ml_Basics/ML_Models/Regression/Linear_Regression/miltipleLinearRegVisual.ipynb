```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
import plotly.express as px
import plotly.graph_objects as go
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Generate DataSet

X,y = make_regression(n_samples=200, n_features=2, n_informative=2, n_targets=1, noise=75)

# Generate df to plot graph

df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
lr = LinearRegression()

lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

```python
# Visualization of Plane

# Define the grid ranges
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
xGrid, yGrid = np.meshgrid(y, x)

#  Create 'final' FIRST (Define the input data)
# .ravel() flattens the 10x10 grids into 100-length arrays
final = np.vstack((xGrid.ravel(), yGrid.ravel())).T

# Predict using 'final' (Now the variable exists)
z_final = lr.predict(final).reshape(10, 10)

# Assign to z for plotting
z = z_final
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.add_trace(go.Surface(x = x, y = y, z =z ))

fig.show()
```

```python
print('Coefficients - ',lr.coef_)

print('InterCepts', lr.intercept_)
```
