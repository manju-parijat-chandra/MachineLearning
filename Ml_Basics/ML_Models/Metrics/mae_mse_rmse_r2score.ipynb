```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df = pd.read_csv('../../../DataBox/titanic_mmimp.csv')

df.dropna(inplace=True)
```

```python
# IO
X = df.iloc[:,0:1]
y = df.iloc[:,-1]

# TTS
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)

lr = LinearRegression()
lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)
```

Calculate Metrics -> MAE , MSE, RMSE, R2Score, Adj_R2Score

```python
# we just have to provide y test (Y) and y predicted (Y_hat) to metric Functions
# How Good is out prediction on testing Data

# Mean Absolute Error
mae = mean_absolute_error(y_test,y_pred)

# Mean Squared Error
mse = mean_squared_error(y_test,y_pred)

# Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test,y_pred))

# R2 Score
r2scr = r2_score(y_test,y_pred)

# Adjusted R2 Score

# Evaluating Training Set
# n = len(X_train)
# k = X_train.shape[1]

# Evaluating Test Set
n = len(X_test)
k = X_test.shape[1]

print('Number of Input Rows', n)
print('Number Of Input Columns', k)

ar2scr = 1 - ((1-r2scr)*(n-1)/(n-1-k))
```

```python
print('MAE - ', mae, '\nMSE - ', mse, '\nRMSE - ', rmse, '\nR2-Score - ', r2scr, '\nAdj R2-Score - ', ar2scr)
```

```python
# MAE -  0.48 -> 48 % Error in our Prediction
# MSE -  0.25 -> Unit is Squared
# RMSE -  0.50 -> Same unit as MAE
# R2-Score -  -0.01
# -> Should not be -ve
# -> Wrong Model is used or did not understood data
# -> not able to capture Variance of Data Point At all
# Adj R2-Score -  -0.02 -> Column used is poorly related
```
