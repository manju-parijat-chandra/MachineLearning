```python
####################   EDA   #########################

# ***************** Univariate Analysis ***************

df = pd.read_csv('../DataBox/titanic.csv',index_col=0) # Passenger ID -> Index

df.drop(columns=['boat','body','home.dest'], inplace=True) # Column with NaN Values

# Survival is Our Output So, We keep such Column that can affect Survival
# Analysis on Categorical Column  
# -> survived(0,1), pclass(1,2,3), Name(names), sex(m,f), sibsp(), parch(), ticket(too many classes), cabin, embarked

# Analysis on Numerical Data
# -> Age (Continuous Data)
```

```python
# --------- Categorical Data -------------- -> PIE Chart , Bar Graph 
            # Survived
fig, ax = plt.subplots(1, 2, figsize=(12, 5))            
            
# Count Plot -> Frequency Count -> Distribution
# sns.countplot(df['survived'])

df['survived'].value_counts().plot(kind='bar', ax=ax[0])


# PIE Chart -> Percentage Distribution

df['survived'].value_counts().plot(kind='pie', autopct='%.2f%%', ax=ax[1])

plt.show()
```

```python
             # Pclass
fig, ax = plt.subplots(1, 2, figsize=(12, 5))            
# Frequency Distribution
df['pclass'].value_counts().plot(kind='bar', ax=ax[0])  # Class Distribution

# Percentage Distribution
df['pclass'].value_counts().plot(kind='pie', autopct='%.2f%%', ax=ax[1])

plt.show()
```

```python
#                     Numerical Data 
fig, ax = plt.subplots(1, 2, figsize=(12, 5)) 
#                       Age    (As Categorical Data if added Bins)

#  ******************** Histogram -> To understand Distribution *********************
ax[0].hist(df['age'], bins=30)

# Histogram with Probability Density Function
sns.displot(df['age'], kde=True, stat="density") #bThis also Tell us About Skewness Of Data

# 5 Number Summary (Box Plot) -> Check Noise
sns.boxplot(df['age'], ax=ax[1])

stats = df['age'].agg(['max', 'min', 'mean', 'std', 'skew'])
print(stats)
```

```python
# ***************** Bivariate / Multivariate Analysis ***************
# Analysis of Two or More Columns Together 
# Categorical - Categorical , Categorical - Numerical, Numerical - Numerical


tips = sns.load_dataset('tips')

flights = sns.load_dataset('flights')

iris = sns.load_dataset('iris')

titanic = pd.read_csv('../DataBox/titanic.csv',index_col=0)
```

```python
# ****************** Scatter Plot *********************

# Numerical - Numerical
# **** 2 Columns (Bivariate) ****
sns.scatterplot(data=tips, x='total_bill', y='tip') # Check what is the Trend


# **** Multi variate ****
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex',style='smoker', size='size') # 5 Columns (Information in one Graph)
```

```python
# ************* Bar Plot ***********
# Numerical - Categorical


# Bivariate Analysis
sns.barplot(data=titanic, x='pclass', y='fare')  #--> Confidence Interval is Included

# Multivariate Analysis
sns.barplot(data=titanic, x='pclass', y='age', hue='sex') 



# *************  Box Plot  **************** For Bivariate and Multivariate Analysis
# Numerical - Categorical

sns.boxplot(data=titanic, x='sex', y='age', hue='survived')



# ************* Distplot  **************   (Numerical and Categorical)

sns.displot(data=titanic, x = 'age', hue = 'survived', kind = 'kde')

```

```python
# ---------- Heat Map ------------  (Categorical - Categorical)

# cross-tabulation -> In Numbers
ctab = pd.crosstab(titanic['pclass'], titanic['survived'], margins=True)

# HeatMap Plotting
sns.heatmap(ctab)


# In Terms Of Percentage
(titanic.groupby('pclass').mean(numeric_only=True)['survived']*100).plot(kind='bar') # numeric_only needed  -> Pandas 2.0+

titanic.groupby('embarked').mean(numeric_only=True)['survived']*100



# ---------- Cluster Map ------------  (Categorical - Categorical)

cluster = pd.crosstab(titanic['parch'], titanic['survived'])

sns.clustermap(cluster) # Dendrogram
```

```python
# ---------- Pairplot ------------  Collection Of Scatter Plot (Numerical - Numerical)
# Automatically Detect Numerical Column and Present Relation BTW Them 

sns.pairplot(iris) # Histogram and Scatter Plot

# Multicariate analysis
sns.pairplot(iris, hue='species')
```

```python
# ---------- LinePlot ------------  Special Case Of Scatter Plot -> Time-Based Data -> (Numerical - Numerical)

new_db = flights.groupby('year').sum(numeric_only=True).reset_index()

sns.lineplot(data= new_db, x='year', y='passengers')
```

```python
# -------- Pivot Table ----------- (Numerical - numerical)

piv_tab = flights.pivot_table(values='passengers',index='month',columns='year')

sns.heatmap(piv_tab)

sns.clustermap(piv_tab)
```
