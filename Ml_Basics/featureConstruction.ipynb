```python
# Two Method -> Feature Splitting and Generatre new Feature

# Imports

import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv('../DataBox/titanic_prepsd.csv')[['Age','Pclass','SibSp','Parch','Survived']]

df.dropna(inplace=True)
```

```python
# Input output
X = df.iloc[:,0:4]
y = df.iloc[:,-1]

# just check the Score
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

Apply Feature Construction (Generate New Column family size)

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1 # +1 for the passenger

# Return Family Classification
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2

X['Family_type'] = X['Family_size'].apply(myfunc)

# Drop Extra Columns (Old Column)
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)

# Calculate New Score on the basis of changes (New Column)
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))

# Observation 
# Score Improved
```

Feature Splitting

```python
df = pd.read_csv('../DataBox/titanic_prepsd.csv')[['Name','Age','Pclass','SibSp','Parch','Survived']]

df.dropna(inplace=True)
```

```python
# Input output
X = df.iloc[:,1:5]
y = df.iloc[:,-1]

# just check the Score
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

```python
# Split Name into Two column -> Title and Name

df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0] # Salutation

df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]


# Analysis

(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)

# Observation -> Mrs - 0.792000 and Miss - 0.697802  *** Has More Survival Rate ***
# we can use this info to generate new columns
```

```python
# Create new column with Is_Married.

df['Is_Married'] = 0

# Assign 1 if married
df.loc[df['Title'] == 'Mrs', 'Is_Married'] = 1

# Drop Extra Columns (Old Column)
df.drop(columns=['Title','Name'],inplace=True)

# Input output
X = df[['Age', 'Pclass', 'SibSp', 'Parch', 'Is_Married']]
y = df['Survived']

# just check the Score
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))

# observation 
# Score increased From 0.693 to 0.756
```

```python
# Create two new columns with Married_Women and Unmarried_Girl.

df['Married_Women'] = 0
df['Unmarried_Girl'] = 0

# Assign 1 if married or unmarried 
df.loc[df['Title'] == 'Mrs', 'Married_Women'] = 1
df.loc[df['Title'] == 'Miss', 'Unmarried_Girl'] = 1

# Drop Extra Columns (Old Column)
df.drop(columns=['Title','Name'],inplace=True)

# Input output
X = df[['Age', 'Pclass', 'SibSp', 'Parch', 'Married_Women','Unmarried_Girl']]
y = df['Survived']

# just check the Score
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))

# observation 
# Score increased From 0.756 to 0.811
```
