```python
# Imports 
import numpy as np
```

```python
## *************** VECTORS *****************

# Define N-dimentional Vector

A = np.array([1,2,3,4,5,6]) # 6 Components of Vector A  ->  A is 6D Vector

# Types of Norms (Magnitude, Size, Length)
# L0 Norm -> Sparsity -> Count of non-Zeros
# L1 Norm -> Manhattan -> Sum of Absolutes
# L2 Noem -> Euclidean -> Square root of sum of Squares
# L infinity -> Maximum -> Mas absolute value

# Calculate Euclidean Distance from Origin (L2 Norms)

distance = np.linalg.norm(A)

print("Euclidean distance of Vector A is : ", distance)
```
