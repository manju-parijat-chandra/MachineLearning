```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import euclidean
```

```python
## *************** VECTORS *****************

# Define N-dimentional Vector

A = np.array([1,2,3,4,5,6]) # 6 Components of Vector A  ->  A is 6D Vector

# Types of Norms (Magnitude, Size, Length)
# L0 Norm -> Sparsity -> Count of non-Zeros
# L1 Norm -> Manhattan -> Sum of Absolutes
# L2 Noem -> Euclidean -> Square root of sum of Squares
# L infinity -> Maximum -> Mas absolute value

# Calculate Euclidean Distance from Origin (L2 Norms) -> (Size of Vector)

distance = np.linalg.norm(A)

print("Euclidean distance of Vector A : ", distance)

# Euclidean Vector Between TWO Vector -> (Distance between two point)

B = np.array([1,2,3,4,6,6])
# Find A - B then -> Euclidean Distance of vector (A-B) from the Origin

difference = A - B

# Length of vector difference 

distance = np.linalg.norm(difference)

print("Euclidean distance of Vector A - B (or distance between vector A and B or distance btw point A and B) : ", distance)


#     **** Important Observation ****
# If vector A = [1,2,3,4,5,6]
# If vector B = [1,2,3,4,5,7]   -- Or -- [1,2,3,4,6,6]
# We can bring them to origin point [1,2,3,4,5,6]
# Then Vector A will become -> Origin 
# And Vector B Will become -> [0,1] -- Or -- [1,0]
```

```python
# ************* Plotting On Graph ****************

# Generate 5 random ** 3D ** vectors
vectors = np.random.rand(5,3) # Matrix of 5 rows and 3 column

# Assign random class (0 or 1) to each Vector ('Categorization') -> Adding Cotegory Column at index -> 0.
# New Matrix has 5 rows and 4 column

classes = np.random.randint(0,2,5)

# Plot vectors on #D Matplotlib Graph

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for vec, cls in zip(vectors, classes):
    ax.quiver(0,0,0, vec[0], vec[1], vec[2], color=('r' if cls == 1 else 'b'))


# Get Input Vector -> New Point has appeared
query_vector = np.array(list(map(float, input("Enter a point(3D Vector) : ").split())))

# Calculate the distance of new vector from all 5 vectors
distance = [euclidean(query_vector, vec) for vec in vectors]
nearest_neighbor_index = np.argmin(distance)

# Output the class of nearest neighbour

print("class according to nearest neighbour is : ", classes[nearest_neighbor_index])

# plot the new vector
ax.quiver(0,0,0, query_vector[0], query_vector[1], query_vector[2], color='g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

```python
# Mean Centering (Bring Origin to Mean)

# Generate 100 - 2D Vectors

data = np.random.rand(100,2)

# Plot original data

plt.subplot(1,2,1)
plt.scatter(data[:, 0], data[:, 1], label="Original Data")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Before mean Centering")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
```

```python
# Calculate the mean of data along each dimension

data_mean = np.mean(data, axis=0)
# data_mean -> [x_mean, y_mean] (Matrix with 2 column -> mean of x and y axis)

# Perform Mean centerin by Shifting the Origin

cantered_data = data - data_mean # Origin is Shifted to mean of X,Y

plt.subplot(1,2,1)
plt.scatter(cantered_data[:, 0], cantered_data[:, 1], label="Mean Centered Data")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("After mean Centering")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
```

```python
# Dot Product A.B

# Define two Matrix (n x 1)
A = np.array([1,2,3])
B = np.array([4,5,6])
C = np.array([7,8,9])

# First It traspose Matrix A then Does Matrix multiplication -> A [n x 1] --> A [1 x n] then (A [1 x n]) * (B [n x 1])
dot_Product = np.dot(A,B)

dot_product_new_way = A@B

print("Traditional : ", dot_Product)
print("New : ", dot_product_new_way)

# Properties of . product

# 1->>> Commutative Property  ->  A.B = B.A

dot_Product_AB = np.dot(A,B)
dot_product_BA = np.dot(B,A)

print("Here A.B : ", dot_Product_AB, "->  B. A : ", dot_product_BA)

# 2->>> Distributive Property  ->  A.(B+C) = A.B + A.C

dot_product_A_BC = np.dot(A, B+C)
dot_Product_AB_plus_BC = np.dot(A,B) + np.dot(A,C)

print("Here A . (B + C) : ", dot_product_A_BC, " ->  (A . B) + (A . C) : ", dot_Product_AB_plus_BC)
```

```python
# Use case of Dot Product
# Cosine Similarity between vectors  ->  cos(Q) = A.B / ||A|| * ||B||

cosine_similarity_AB = np.dot(A,B) / (np.linalg.norm(A)*np.linalg.norm(B))
print("Cosine Similarity BTW A,B : ", cosine_similarity_AB)

cosine_similarity_AC = np.dot(A,C) / (np.linalg.norm(A)*np.linalg.norm(C))
print("Cosine Similarity BTW A,C : ", cosine_similarity_AC)
```
