```python
import pandas as pd
import numpy as np
from scipy import stats
import random
```

```python
# ******* Probability Distribution function ********

#  -> Probability Mass Function

# 1 Dice is Rolled at a time
L = []
for i in range (10000):
    # Storing random number between 1-6 (Sample Space - s)
    L.append(random.randint(1,6))
```

```python
# Count of an Event / Total Number of Event
s = (pd.Series(L).value_counts() / pd.Series(L).value_counts().sum()).sort_index()
s.plot(kind='bar')

# Uniform Probability Distribution
# Probabilities are same
```

```python
# 2 Dice is Rolled at a time and Sum of numbers is our outcome
X = []
for i in range (10000):
    # Storing random number between 1-6 (Sample Space - s)
    a = random.randint(1,6)
    b = random.randint(1,6)
    
    X.append(a+b)
```

```python
# Count of an Event / Total Number of Event
s = (pd.Series(X).value_counts() / pd.Series(X).value_counts().sum()).sort_index()
s.plot(kind='bar')

# Probabilities are different
```

```python
# Cumulative Distribution Function of PMF

# PMF ---> Gives probability of a particular point x
# CDF ---> Gives probability upto a particular point x  (i.e-> <= x)

cdf = np.cumsum(s)
cdf.plot(kind='bar')
```
